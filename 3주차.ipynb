반복문과 조건문을 함께 사용

```python
scores = [42 ,60, 77, 89, 20]
for score in scores:
    if score >= 60:
        print(f'{score}점은 합격입니다')
    else:
        print(f'{score}점은 불합격입니다')

        #60점은 합격입니다
        # 42점은 불합격입니다   
```

```python
numbers = [1,2,3,4,5]
#합계: numbers를 다 더한 값
total = 0
for number in numbers:
    total += number
print("합계:",total)
```

```python
scores = [65,82,91,100,58,77]
#70점 이상은 합격
#합격자는 x명 입니다
total = 0
for score in scores:    
    if score >=70 :   
        total += 1
print(f'합격자는 {total}명 입니다')
    
```
